```python
# pip install -e .
import pandas as pd
import numpy as np

import stfind as sf
from stfind.buildkernel import spatialkernel, isPD, buildchol, nearestPD
```

```python
dir="../data/"
```

```python
exprs = pd.read_csv(f'{dir}/example_exp.csv.gz', index_col=0)
exprs
```

```python
coord = pd.read_csv(f'{dir}/example_location.csv.gz', index_col=0)
coord
```

```python
skernel = spatialkernel(np.array(exprs), np.array(coord), bandwidthtype='silverman')
skernel
```

```python
isPD(skernel)
```

```python
skernel_nearPD = nearestPD(skernel)
skernel_nearPD
```

```python
chol_skernel = np.linalg.cholesky(skernel_nearPD)
chol_skernel
```

```python
ori_skernel = chol_skernel @ chol_skernel.T
ori_skernel
```

```python
np.round(skernel_nearPD, 6) == np.round(ori_skernel, 6)
```
